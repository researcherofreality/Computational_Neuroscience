```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def binom(n, k):
    return math.factorial(n) // math.factorial(k) // math.factorial(n - k)

def L(ph):
  return binom(N,heads)*(ph**heads)*(1-ph)**(N-heads)
```

```python
heads = 19
N = 30
ph = np.linspace(0,1,100)
L(ph);
```

```python
plt.plot(ph,L(ph));
plt.xlabel('pH')
plt.ylabel('likelihood')
plt.title('the likelihood as a function of pH, given heads = 19 & N = 30')
plt.show()
```

```python
import scipy.integrate as intg
res, err = intg.quad(L,0,1)
print("the integral over the likelihood is {:.3f} +- {:.3f}".format(res,err))
print("the maximum likelihood estimate for ph is {:.3f} at ph={:.3f}".
      format(np.max(L(ph)),ph[np.argmax(L(ph))]))
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Nsteps = 1000
positions = np.zeros((Nsteps+1,2))

for i in range(Nsteps):
  positions[i+1][0] = positions[i][0] + np.random.randn()
  positions[i+1][1] = positions[i][1] + np.random.randn()

x=positions[:,0]
y=positions[:,1]
sqrdisp = x**2 + y**2

plt.plot(x,y)
plt.plot(0,0,'ro')
plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel('x-position')
plt.ylabel('y-position')
plt.title('2D random walk for N = 1000, sigma = 1.0')
plt.show()
```

```python
plt.plot(range(Nsteps+1),sqrdisp)
plt.xlabel('t')
plt.ylabel('squared displacement')
plt.title('squared displacement as a function of time, N = 1000')
plt.show()
```

```python
sigmas = [0.2, 0.5, 1.0, 1.5]

for s in sigmas:
  dispmat = []
  runs = 1000
  for j in range(runs):
    Nsteps = 1000
    positions = np.zeros((Nsteps+1,2))

    for i in range(Nsteps):
      positions[i+1][0] = positions[i][0] + s*np.random.randn()
      positions[i+1][1] = positions[i][1] + s*np.random.randn()

    x=positions[:,0]
    y=positions[:,1]
    sqrdisp = x**2 + y**2
    dispmat.append(sqrdisp)

  meansqrdisp = np.mean(dispmat,axis=0)
  plt.plot(range(Nsteps+1),meansqrdisp)

plt.legend(sigmas)
plt.xlabel('t')
plt.ylabel('MSD')
plt.title('MSD as a function of time, varying sigma')
plt.show()
```
