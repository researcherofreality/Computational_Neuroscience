# Deriving and defining velocity and acceleration function

```python
# take derivative of xt and yt w.r.t. tau
x0,xF,t,T = symbols('x0,xF,t,T')
xt = x0 + (x0-xF)*(15*(t/T)**4-6*(t/T)**5-10*(t/T)**3)
dxt = xt.diff(t)
dxt2 = dxt.diff(t)
print(dxt)

#same for y
y0,yF,t,T = symbols('y0,yF,t,T')
yt = y0 + (y0-yF)*(15*(t/T)**4-6*(t/T)**5-10*(t/T)**3)
dyt = yt.diff(t)
dyt2 = dyt.diff(t)
print(dyt)

v = sqrt(dxt**2+dyt**2)
a = sqrt(dxt2**2+dyt**2)
print('v={}'.format(v))
print('a={}'.format(a))
```

```python
def v(x0,y0,xF,yF,T):
  t = np.linspace(0,T,100)
  tau = t/T
  return tau, np.sqrt((x0 - xF)**2*(-30*t**2/T**3 + 60*t**3/T**4 - 30*t**4/T**5)**2 + (y0 - yF)**2*(-30*t**2/T**3 + 60*t**3/T**4 - 30*t**4/T**5)**2)
```

```python
def a(x0,y0,xF,yF,T):
  t = np.linspace(0,T,100)
  tau = t/T
  return tau, np.sqrt((x0 - xF)**2*(-60*t/T**3 + 180*t**2/T**4 - 120*t**3/T**5)**2 + (y0 - yF)**2*(-30*t**2/T**3 + 60*t**3/T**4 - 30*t**4/T**5)**2)
```

# Plotting velocity profiles

```python
tvals,v_d = v(0,0,0.1,0.1,1)

plt.plot(tvals,v_d)
plt.title('velocity profile of the hand movement')
plt.xlabel('t')
plt.ylabel('velocity (m/s)')
plt.show()
```

```python
# space scaling: increasing the distance to be travelled
labels = []
for i in [0.1,0.5,1]:
  tau,v_d = v(0,0,i,i,1)
  plt.plot(tau,v_d)
  labels.append('{}m'.format(i))

plt.title('velocity profiles for varying distance to be travelled'.format(labels))
plt.legend(labels)
plt.show()
```

```python
# time scaling: same distance w different travel times
labels = []
for i in [1,2,5]:
  tau,v_d = v(0,0,0.1,0.1,i)
  plt.plot(tau,v_d)
  labels.append('{}s'.format(i))

plt.title('velocity profiles for a distance w different travel times'.format(labels))
plt.legend(labels)
plt.show()
```

# Plotting acceleration profiles

```python
tvals,v_d = a(0,0,0.1,0.1,1)

plt.plot(tvals,v_d)
plt.title('acceleration profile of the hand movement')
plt.xlabel('t')
plt.ylabel('acceleration (m/s^2)')
plt.show()
```

```python
# space scaling: increasing the distance to be travelled
labels = []
for i in [0.1,0.5,1]:
  tau,v_d = a(0,0,i,i,1)
  plt.plot(tau,v_d)
  labels.append('{}m'.format(i))

plt.title('acceleration profiles for varying distance to be travelled'.format(labels))
plt.legend(labels)
plt.show()
```

```python
# time scaling: same distance w different travel times
labels = []
for i in [1,2,5]:
  tau,v_d = a(0,0,0.1,0.1,i)
  plt.plot(tau,v_d)
  labels.append('{}s'.format(i))

plt.title('acceleration profiles for a distance for different travel times'.format(labels))
plt.legend(labels)
plt.show()
```
